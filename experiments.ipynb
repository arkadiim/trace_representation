```python
# SageMath version 7.5.1, Release Date: 2017-01-15
def product(repeat = 1, *args):
    # product('ABCD', 'xy') --> Ax Ay Bx By Cx Cy Dx Dy
    # product(3, range(2)) --> 000 001 010 011 100 101 110 111
    result = [[]]
    for _ in range(repeat):
        for pool in args:
            result = [x + [y] for x in result for y in pool]
#     for prod in result:
#         yield prod
    return result

def majfunc(n):
    # генерирует функцию голосования от n переменных
    tt = []
    for i in product(n, [0, 1]):
        n0 = len(filter(lambda x: x==0, i))
        n1 = len(filter(lambda x: x==1, i))
#         tt.append((''.join(map(str, i)), int(n1>n0)))
        tt.append(int(n1>n0))
    return tt

def Tr2(n, a):
    s = a
    for _ in xrange(n-1):
        a = a ** 2
        s += a
    return s

def cyclotomic_classes2(m):
    '''
    return: dict 
    {[element of Z_m] : ([class generator], [power], [n_s])}
    '''
    l = range(m)
    classes = []
    while len(l) > 0:
        s = l[0]
        cls = []
        while True:
#             print(s, l, cls)
            l.remove(s)
            cls.append(s)
            s = (2*s) % m
            if s == cls[0]: break
        classes.append(cls)
    classes.append([m])
#     print classes
    d = {cls[i]: (cls[0], (len(cls) - i) % len(cls), len(cls))
         for cls in classes for i in xrange(len(cls))}
    d['classes'] = classes
    return d

def find_lambda(elems_of_field):
    n = Integer(len(elems_of_field))
    n = n.exact_log(2)
    for a in elems_of_field:
        if Tr2(n, a) == 1: return a

        
def latex_tr(l):
    string = []
    for deg, b_s, n_s in l:
        if b_s == 0: continue
        if b_s == 1:
            string.append('Tr_{{{}}}(x^{})'.format(n_s, deg))
        else:
            string.append('Tr_{{{}}}({} x^{})'.format(n_s, b_s, deg))
    return ' + '.join(string)
        

from IPython.display import Markdown as md
from IPython.display import display
def Mat(F_list):
    q = len(F_list)
    M = matrix(F, [map(lambda a: a**i, F_list) for i in xrange(q-1, -1, -1)])
    # M = matrix(F, [[a**i for i in xrange(q-1, -1, -1)] for a in F_list])
    # M = M.T
    M[0] = M[0] + vector([F('-1') for _ in range(q)])
    M = -M
#     display(md('$${}$$'.format(latex(M))))
    return M


import itertools
def uni_combinations(q, m):
    '''все последовательности длины q с m единицами'''
    for comb in itertools.combinations(range(q), m):
        seq = [0]*q
        for i in comb: seq[i]=1
        yield seq


    
    
def trace_repr(value_vector, M, q):
    value_vector = vector(F, value_vector)

    f = {i: v for i, v in enumerate(M*value_vector)}
    print '     ', value_vector
#     print list(M*value_vector), f
    print R(f)#, type(R(p))
#     print 'true_poly?', [R(f)(y) for y in F_list]
    # print 'cycl_clss', cyclotomic_classes2(q-1)
    lmd = find_lambda(F_list)
    print 'lmd = ', lmd
    cycl_clss = cyclotomic_classes2(q-1)
    print 'cycl_clss', cycl_clss, cycl_clss['classes']
    f_reduce = {cls[0]: F('0') for cls in cycl_clss['classes']}
    for deg, coef in f.items():
        coef = lmd * coef
        cls_el = cycl_clss[deg]
        f_reduce[ cls_el[0] ] += coef**(2**cls_el[1])
    print 'new_f_reduce', R(f_reduce), f_reduce
#     print [Tr2(n, R(f_reduce)(a)) for a in F_list]

    trace_form = []
    for deg, coef in f_reduce.items():
        b_s = coef
        n_s = cycl_clss[deg][2]
        for i in xrange(n_s, n-n_s+1, n_s):
            coef = coef ** (2**n_s)
            b_s += coef
        trace_form.append((deg, b_s, n_s))
    print 'check', [sum([ Tr2(n_s, b_s * a**deg ) for deg, b_s, n_s in trace_form]) for a in F_list]
#     print trace_form
    print latex_tr(trace_form)
#     display(md('$${}$$'.format(latex_tr(trace_form))))
    count_zero = len(filter(lambda x: True if x == 0 else False, [b_s for _, b_s, _ in trace_form]))
    print 'count_zero', count_zero#, '\n\n'

    return trace_form
```
